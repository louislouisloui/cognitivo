### FIXED ACIDITY

* VARIANCE UNIFORM
* LINEAR

```python
quality = dt.quality.unique()
quality.sort()
plt.figure(figsize=(15,8))
for i in quality:
    sns.kdeplot(dt[dt.quality == i]['fixed acidity'], label = f'quality = {i}')
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(data = dt, x = 'quality', y = 'fixed acidity')
```
